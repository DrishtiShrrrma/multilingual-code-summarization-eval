```python
import os
import json

def explore_directory(root_dir, max_preview_lines=5):
    print(f"\nExploring directory: {root_dir}\n{'='*50}")

    for dirpath, dirnames, filenames in os.walk(root_dir):
        level = dirpath.replace(root_dir, '').count(os.sep)
        indent = ' ' * 4 * level
        print(f"{indent}{os.path.basename(dirpath)}/")
        sub_indent = ' ' * 4 * (level + 1)

        for fname in filenames:
            print(f"{sub_indent}{fname}")
            file_path = os.path.join(dirpath, fname)

            # Preview content for text or json files
            if fname.endswith(".json"):
                try:
                    with open(file_path, 'r', encoding='utf-8') as f:
                        data = json.load(f)
                        print(f"{sub_indent}  Preview (first entry): {json.dumps(data[0], indent=2)[:300]}...\n")
                except Exception as e:
                    print(f"{sub_indent}  [Error reading JSON: {e}]")
            elif fname.endswith(".txt"):
                try:
                    with open(file_path, 'r', encoding='utf-8') as f:
                        for i in range(max_preview_lines):
                            line = f.readline()
                            if not line: break
                            print(f"{sub_indent}  {line.strip()}")
                except Exception as e:
                    print(f"{sub_indent}  [Error reading TXT: {e}]")

# Example usage
explore_directory("nueva/prompt_analysis")
```
